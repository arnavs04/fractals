# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Functions

```python
def mandelbrot(c, max_iter):
    z = 0
    n = 0
    while abs(z) <= 2 and n < max_iter:
        z = z*z + c
        n += 1
    return n
```

```python
def mandelbrot_set(xmin, xmax, ymin, ymax, width, height, max_iter):
    x = np.linspace(xmin, xmax, width)
    y = np.linspace(ymin, ymax, height)
    mandelbrot_grid = np.zeros((height, width))
    for i in range(height):
        for j in range(width):
            mandelbrot_grid[i, j] = mandelbrot(complex(x[j], y[i]), max_iter)
    return mandelbrot_grid
```

```python
def plot_mandelbrot(xmin, xmax, ymin, ymax, width, height, max_iter):
    mandelbrot_grid = mandelbrot_set(xmin, xmax, ymin, ymax, width, height, max_iter)
    plt.figure(dpi = 1200)
    plt.imshow(mandelbrot_grid, extent=(xmin, xmax, ymin, ymax), cmap='magma')
    plt.axis('off')
    plt.show()
```

# Plotting

```python
xmin, xmax = -2, 2
ymin, ymax = -2, 2
width, height = 1000, 1000
max_iter = 100

# Plot the Mandelbrot set
plot_mandelbrot(xmin, xmax, ymin, ymax, width, height, max_iter)
```
