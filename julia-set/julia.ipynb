# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Functions

```python
def julia_set(width, height, cx, cy, max_iter):
    xmin, xmax = -2, 2
    ymin, ymax = -2, 2
    
    x, y = np.meshgrid(np.linspace(xmin, xmax, width), np.linspace(ymin, ymax, height))
    z = x + 1j * y
    
    julia = np.zeros_like(z, dtype=np.float32)
    
    for i in range(max_iter):
        z = z**2 + cx + 1j * cy
        mask = np.abs(z) < 1000
        julia += mask
    
    return julia
```

```python
def plot_julia_set(julia):
    plt.figure(figsize=(8, 8), dpi=2000)
    plt.imshow(julia, cmap='bone', extent=(-2, 2, -2, 2))
    plt.axis('off')
    plt.show()
```

# Plotting

```python
width, height = 800, 600
cx, cy = -0.7, 0.27015
max_iter = 1000
    
julia = julia_set(width, height, cx, cy, max_iter)
plot_julia_set(julia)
```
